```python

import numpy as np
from numpy import linspace

m = 70
g = 9.8

A = 0.05
z0 = 1
T = 0.5
w = 2*np.pi / T
t =  linspace(0,T,100)
z = A * np.sin(w*t + w*T/4.) + z0

B = 0.5
v0 = 5
vx = B * np.sin(w*t + w*T/4.) + v0
vz = w * A * np.sin(w*t + w*T/4.)

V = m*g*z
E = 0.5 * m * (vx**2 + vz**2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
plt.plot(t,V, label='potential energy')
plt.plot(t,E, label='kinetic energy')
plt.legend()
plt.xlabel('time')
plt.ylabel('energy')
plt.show()
```
